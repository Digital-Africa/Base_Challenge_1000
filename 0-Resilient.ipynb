```python
def percent(target, element):
    perc = list(numpy.percentile(target, [50, 75, 90]))
    if element < perc[0]:
        return 0
    if element >= perc[0] and element < perc[1]:
        return 1
    else:
        return 2 
    
recovery['label_age'] = recovery.apply(lambda x: percent(age,todigit(x['Age (entreprise)']) ), axis = 1)
recovery['label_salarie'] = recovery.apply(lambda x: percent(age,todigit(x['nbr_salarie']) ), axis = 1)
recovery['label_croissance'] = recovery.apply(lambda x: percent(age,todigit(x['Croissance 2018-2019']) ), axis = 1)
recovery['label_ca_19'] = recovery.apply(lambda x: percent(age,todigit(x["Chiffre d'affaire 2019"]) ), axis = 1)
#recovery['indicateur_label'] = recovery['label_target_category']*recovery['label_made_in_africa']*(recovery['label_ca_19']+recovery['label_croissance']+recovery['label_salarie']+recovery['label_age'])
#recovery['percent_indicateur_label'] = recovery.apply(lambda x: percent(recovery['indicateur_label'].values.tolist(),todigit(x["indicateur_label"]) ), axis = 1)
#recovery['pf_label'] = recovery['Tech'].map(todigit)*recovery['Innovant'].map(todigit)#*recovery['Coup de coeur'].map(todigit)
#recovery['full_label'] = recovery['pf_label']*recovery['indicateur_label']
#recovery['percent'] = recovery.apply(lambda x: percent(recovery['full_label'].values.tolist(),todigit(x["full_label"]) ), axis = 1)
```

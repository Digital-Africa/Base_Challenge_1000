```python
raw = challenge1000.Native().raw
schema = challenge1000.RefEnv().schema
def completion_rate_from_schema(raw, schema):
    header = list()
    values_na = list()
    completion_rate = dict()
    for element in schema:
        completion_rate[element] = dict()
        for e in schema[element]:
            if e in raw.keys():
                na = raw[e].isna().sum()
                header.append(e)
                values_na.append(na)
                completion_rate[element][e] = (1-na/len(raw))*100
    return completion_rate

completion_rate_from_schema(raw, schema)
```

```python
import pandas
pandas.DataFrame_from_dict(completion_rate)
```

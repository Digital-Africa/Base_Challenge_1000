```python
crps['text'] = crps.agg(sum, axis = 1).map(lambda x: x.lower())
#crps['text']
```

```python
def contains_keyw(element, kwrd):
    if kwrd in element:
        return True
    else:
        return False
    
#crps.apply(lambda x: contains_keyw(x['text'], 'soil'),  axis= 1)
#len(text[text['soil']])
```

```python
target_list = ['soil', 'nutrient', 'satellite', 'imagery', 'crop effi']
def from_kyeword_list(target_list):
    header = target_list + ['keep']
    for k in target_list:
        crps[k] = crps.apply(lambda x: contains_keyw(x['text'], k), axis = 1)
    crps['keep'] = crps[target_list[0]]
    target_list.pop(0)
    for j in target_list:
        crps['keep'] = crps['keep']| crps[j]


    X_1 = crps[crps['keep']][header].join(X['categorie'])
    display(X_1.groupby('categorie').sum())
    display(X_1[target_list].sum())
    return X_1
```

```python
target_list = ['soil','soils', 'nutrient','nutrients', 'satellite', 'productivity','irrigate','irrigation','fertilizers','fertilizer','crop health','crop yield','crops health','crops yield']
a = from_kyeword_list(target_list)
a[a['categorie'].isin(['agriculture','autres','non renseign√©','technologie'])].groupby('categorie').sum()[['keep','soil']]
```
